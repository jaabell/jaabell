## Preface. Correlated random variables and how to de-correlate them (and re-correlate them)

To begin, lets consider two correlated random variables $x_1$ and $x_2$. Furthermore, let's make them normal (Gaussian) with means $\left< x_1 \right> = 1$ and $\left< x_2 \right> = -1$ and covariance matrix
$$
\mathbf{C} = \left[
\begin{array}{rr}
1 & -1\\ -1 & 4
\end{array}
\right]  \qquad \bar{\mathbf{x}} = \left<\mathbf{x}\right> = \left[\begin{array}{r}
1 \\ -1
\end{array}
\right]
$$

The joint probability density function $f_{x_1x_2}(x_1,x_2)$ of these variables looks like

$$
f_{x_1x_2}(x_1,x_2) = 
\dfrac{1}{(2\pi)^2 \det{\mathbf{C}}} 
\exp 
\left( 
    -\dfrac{1}{2} (\mathbf{x} - \bar{\mathbf{x}})^T \mathbf{C}^{-1} (\mathbf{x} - \bar{\mathbf{x}})
\right)
$$

```python
x_bar = array([[1],[-1]])
C = array([[4,-1],[-1,1]])
Cm1 = inv(C)
x1 = linspace(-5,5,200)
x2 = linspace(-5,5,200)
X1,X2 = meshgrid(x1,x2)
a = X1 - x_bar[0]
b = X2 - x_bar[1]
f_x1x2 = 1./(4*pi**2*det(C)) *exp(-(a**2*Cm1[0,0] + 2*a*b*Cm1[0,1] + b**2*Cm1[1,1])/2)

#plotting
from mpl_toolkits.mplot3d.axes3d import Axes3D
figsize(14,8)
fig = figure()
ax = fig.add_subplot(1,2,1,projection='3d')
#ax.contour(X1,X2,f_x1x2,50)
p = ax.plot_surface(X1,X2,f_x1x2,rstride=1,cstride=4, cmap=cm.jet, linewidth=0.1, antialiased=True)
xlabel("$x_1$")
ylabel("$x_2$")
ax = fig.add_subplot(1,2,2)
clevels = 0.001*array([1e-14,1e-9,1e-6,1e-3, 0.1, 1,2,3,4,5,6,7,8])
p = ax.contour(X1,X2,f_x1x2,clevels,cmap=cm.jet, linewidths=2, antialiased=True)
xlabel("$x_1$")
ylabel("$x_2$")
```

Notice that if we view the distribution from a different coordinate system that is rotated so that the above (right) ellipses are horizonal, we are viewing a uncorrelated set of (normal) random variables. Rotating the coordinate system is a *coordinate transformation*, or change of variable. In this case this is called de-correlation of the random variables. 

It turns out that finding the axes $(x^*_1,x^*_2)$ that make the variables look un-correlated is related to an eigenvalue problem. Indeed the eigen-vectors of the inverse correlation matrix give the directions of these axes.

$$
\mathbf{C} \mathbf{v}_i = \sigma^2_i \mathbf{v}_i \qquad \text{for}\,\, i = 1,2
$$

Where the $i$-th eigenvector $\mathbf{v}_i$ points in the $i$-th principal direction of variance and $\sigma^2_i$ is the squared variance in that direction.

```python
sigma2, v = eig(C)
fig = figure(figsize(9,8))
ax = fig.add_subplot(1,1,1)
p = ax.contour(X1,X2,f_x1x2,clevels,cmap=cm.jet, linewidths=2, antialiased=True)
ax.arrow(x_bar[0,0],x_bar[1,0],v[0,0],v[1,0],width=0.01,color='r')
ax.arrow(x_bar[0,0],x_bar[1,0],v[0,1],v[1,1],width=0.01,color='g')
ax.text(x_bar[0,0] + v[0,0], x_bar[1,0] + v[1,0], "$x^*_1$", color = 'r', size=30)
ax.text(x_bar[0,0] + v[0,1], x_bar[1,0] + v[1,1], "$x^*_2$", color = 'g', size=30)
xlabel("$x_1$")
ylabel("$x_2$")
```

Above, $x_1^*$ is the principal axis of variance (most variance) while $x_2^*$  is the axis of least variance. Once more, note that in these directions the random variables appear uncorrelated (no cross-covariance).

We can use this to simulate the random variables (There is another, more efficient, way which involves Cholesky's decomposition of $\mathbf{C}$ but misses the point i want to make further on). Given $\xi_1$ and $\xi_2$, un-correlated unit variance random variables we can use the eigenvectors $\mathbf{v}_i$ (columns of `v` array in the code) and the variances to re-correlate these random variables to generate the sought ones.

By the way, the variances in each direction in the new axes are given by the eigenvalues of the problem.

```python
print "sigma2_1 = {}, sigma2_2 = {}".format(sigma2[0],sigma2[1])
```

To generate the trials, we create realizations of independent unit Gaussian random variables $\xi_1$ and $\xi_2$, collecting them in vectors $\xi_1^{\text{trials}}$ and $\xi_2^{\text{trials}}$

```python
Ntrials = 200
xi1_trials = randn(Ntrials)
xi2_trials = randn(Ntrials)
figsize(6,6)
scatter(xi1_trials,xi2_trials)
```

Nice! Now form the new variables by projecting with the above eigenvectors and the standard-deviations (remember we have the  variances up there), and then offsetting by the means. 

$$
\mathbf{x}^{\text{trial}} =
\left[
    \begin{array}{c}
        x_1^{\text{trial}} \\ x_2^{\text{trial}}
    \end{array}
\right] = 
\left[
    \begin{array}{cc}
        \sigma_1 & 0 \\ 0 & \sigma_2
    \end{array} 
\right]
\mathbf{v}
\left[ \begin{array}{c}
        \xi_1^{\text{trial}} \\ \xi_2^{\text{trial}}
    \end{array}
\right] + \bar{\mathbf{x}}
$$
We can plot these trials along with our previous countour plots of the PDFs.

```python
sigma = sqrt(sigma2)
x1_trials = sigma[0]*v[0,0]*xi1_trials + sigma[1]*v[0,1]*xi2_trials + x_bar[0,0]
x2_trials = sigma[0]*v[1,0]*xi1_trials + sigma[1]*v[1,1]*xi2_trials + x_bar[1,0]
scatter(x1_trials,x2_trials)
contour(X1,X2,f_x1x2,clevels,cmap=cm.jet, linewidths=1, antialiased=True)
```

It is important to check whether these variables have the statistics we seek. Check the ensemble means and variances.

```python
x1_ensemble_mean = mean(x1_trials)
x2_ensemble_mean = mean(x2_trials)
x1_ensemble_sigma = var(x1_trials, ddof=1)  #ddof = 1 because we're computing variance from samples
x2_ensemble_sigma = var(x2_trials, ddof=1)

print "Simulated x_1 mean: {}   sigma: {}".format(x1_ensemble_mean, x1_ensemble_sigma)
print "Simulated x_2 mean: {}   sigma: {}".format(x2_ensemble_mean, x2_ensemble_sigma)
```

Recall the means are supposed to be 1 and -1 while the variances were 4 and 1. 

Also note that, due to the orthogonality property of eigenvectors, we can recombine the eigenvalues and vectors to form $C$ again by

$$
C = \mathbf{v}^T \mathbf{\Sigma} \mathbf{v} \qquad \text{with}\,\, 
\mathbf{\Sigma} = 
\left[
    \begin{array}{cc}
        \sigma_1^2 & 0 \\ 0 & \sigma_2^2
    \end{array}
\right]
$$

```python
SIG = matrix(v).T*matrix(C)*matrix(v)
Cnew = matrix(v)*matrix(diag(sigma2))*matrix(v).T
print C - Cnew
```

## The KL expansion

Now if you have $N$ correlated random variables arranged in a vector, we can no longer plot these for $N>3$ but the procedure to generate them is the same. Given that we have means and a covariance matrix which will be an $N\times N$ symmetric positive definite matrix. 

Such a vector might represent a sampled *stochastic process*, that is a continuously indexed collection of correlated random variables. Continuously indexed just means that one uses a continuous variable (e.g. time) to identify points (the indexing set). Stochastic processes may be thought like *random functions*, i.e. for each value of $t$ the value of $f(t)$ is random but possibly correlated to neighboring values. 

In the case of a random vector we had a correlation *matrix* giving us the cross-covariance between two components. For stochastic processes we (might) have a correlation *function* that correlates two points.

Lets say we have a second-order stationary stochastic process $w(t)$. This means that the covariance function $c(t_1,t_2)$ exists and dependes only on $\tau = t_2 - t_1$. In the previous section we saw that we could *simulate* a random vector by generating unit normal random variables and combining them using the information gained in an eigenvalue analysis. In other words, we could *expand* the correlated random vector into a sequence of un-correlated random vectors. KL explansion solves the same problem when you have a continuous indexing set as oposed to a discrete one.

KL expansion claims that $w(t)$ can be expanded into

$$
w(t) = \sum_{i=1}^{\infty} \sigma_i v_i(t) \xi_i + \bar{w}(t) 
$$

Here I follow closely the notation in the previous section, $\xi_i$ is an countably infinite set of independent zero mean unit variance random variables, and $v_i(t)$ are corresponding functions that act as eigenvectors of the expansion (they are eigenfunctions :}). $\bar{w}(t) = \left< w(t) \right>$ is a mean function that just reflects the trend in $w(t)$. Finally, $\sigma_i$ is the standard-deviation associated with the $i$-th component. Since we use computers to do real life work this expansion will be truncated after, say,  $N$ terms. Evaluating the truncation error becomes an issue which we kindly forward to our mathematician friends, lets just say that KL expansion is the expansion out of all the possible expansions that minimizes the truncation error.

Anyway, the associated eigenvalue/eigenfunction problem is

$$
\int_D C(t - \tau) v_i(\tau) \,\mathrm{d} \tau = \sigma_i^2 v_i(t)
$$

Which `#ThanksToMaths` we know that it is a [homogeneous Fredholm integral equation of the second kind](http://en.wikipedia.org/wiki/Fredholm_integral_equation). Let's say that $\bar{w}(t) = t$ and $C(\tau) = \exp \left\lbrace  -\vert \tau \vert / b\right\rbrace$ for $t \in [-a,\, a]$. The parameter $b$ is called the correlation time because it can be interpreted as the time interval for which the variables are correlated significantly. Such a model correlation function (or kernel function in Fredholm mathy-talk) represents Markov processes. The function looks like this.

```python
b = 1.
a = 0.5
c = 1./b
tau = linspace(-a, a, 1000)
t = linspace(-a, a, 1000)
cov = exp(-abs(tau)/b)
figsize(10,4)
plot(tau, cov, linewidth = 2)
xlabel(r'$\tau$')
ylabel(r"$c(\tau)$")
```

There is a closed-form solution to this equation that can be found in the Ghanem and Spanos book. I will turn my attention to the numerical solution of such an equation. 

Take the eigen-value problem and multiply it by $\delta v_j(t) \,\mathrm{d}t$ and integrate over the domain $D$. This follows usual engineering approach of using virtual displacements (variational) approach to come up with a weak-form solution to the problem. 

If we express $v_i(\tau) = v_i^j H_j(\tau)$ (assuming summation convention) we get
$$
\int_D C(t - \tau) v_i^k H_j(\tau) H_k(t) \,\mathrm{d} \tau \mathrm{d} t = \sigma_i^2 v_i^k H_j(t)  H_k(t)
$$
Now, pre-multiply by 

$$
\int_D \int_D C(t - \tau) \delta v_j(t) v_i(\tau) \,\mathrm{d} \tau \mathrm{d} t= \sigma_i^2 \int_D   \delta v_j(t) v_i(t)\,\mathrm{d} t 
$$

Using this function (read Ghanem and Spanos's book) the process can be expanded into.

$$
w(t) = t + \sum_{i = 1}^{\infty}\Big[ \sigma_i v_i(t) \xi_i + \sigma_i^* v_i^*(t) \xi_i^* \Big]
$$

Where, due to a technicism in the solution process, we have two countably infinite sets of eigen-values/functions instead of one (which is a countably infinite set of functions anyway). Defining $c = 1/b$

$$
\begin{array}{cc}
    \sigma_i^2 = \dfrac{2c}{\omega_i^2 + c^2} & 
    v_i(t) = \dfrac{\cos (\omega_i t)}{ \sqrt{ a + \dfrac{\sin (2\omega_i a)}{2 \omega_i }} } \\
    \sigma_i^2 = \dfrac{2c}{\omega_i^{*2} + c^2} & 
    v_i^*(t) = \dfrac{\sin (\omega_i^* t)}{ \sqrt{ a - \dfrac{\sin (2\omega_i^* a)}{2 \omega_i^* }} } 
\end{array}
$$

And the pairs $(\omega_i,\,\omega^*_i)$ are the $i$-th solutions to the nonlinear equations

$$
f_1(\omega) = c - \omega \tan (\omega a) = 0 \\
f_2(\omega^*) = \omega^* - c \tan (\omega^* a) = 0 
$$

Lets compute a bunch of these values 

```python
#==== Begin copy-paste code from the web!==================================================================
# From http://stackoverflow.com/questions/13054758/python-finding-multiple-roots-of-nonlinear-equation
# With modifications that supress print output and use brentq instead of custom bisection method
# Also added comments

#Strategy: subdivide [a,b] into subintervals of length dx and use brentq within each interval
from scipy.optimize import brentq

def rootsearch(f,a,b,dx): 
    #Gets new sub-interval, returns None on exit.
    #Ensures there is a sign change in the returned sub-interval
    #if there's no sign change, advance to next interval
    x1 = a; f1 = f(a)
    x2 = a + dx; f2 = f(x2)
    while f1*f2 > 0.0:
        if x1 >= b:
            return None,None
        x1 = x2; f1 = f2
        x2 = x1 + dx; f2 = f(x2)
    return x1,x2

def roots(f, a, b, eps=1e-6):
    #Main root finder
    rootlist = []
    while 1:
        x1,x2 = rootsearch(f,a,b,eps) 
        if x1 != None:
            a = x2
            root = brentq(f,x1,x2)
            if root != None:
                pass
                rootlist.append(root)
        else:
            return rootlist
#==== End copy-paste code from the web!==================================================================
```

```python
#Search for roots within this range
wmin = 0
wmax = 32*c
dw = wmax - wmin
Nsubintervals = 400

f1=lambda w:c - w*tan(w*a)
f2=lambda w:w - c*tan(w*a)
w_list_ = roots(f1, wmin, wmax, eps = dw/Nsubintervals)
wstar_list_ = roots(f2, wmin, wmax, eps = dw/Nsubintervals)

#Remove spurious zeros where the function is infinite
tol = 1e-2
w_list = []
wstar_list = []
for w in w_list_:
    if abs(f1(w)) > tol:
        pass
    else:
        w_list.append(w)
for w in wstar_list_:
    if abs(f2(w)) > tol or abs(w) < tol:
        pass
    else:
        wstar_list.append(w)
w_list = array(w_list)
wstar_list = array(wstar_list)

#Plotting
print "Found {} roots for w in [{},{}]".format(len(w_list) + len(wstar_list), wmin, wmax)
fig = figure(figsize(20,4))
w = linspace(wmin,wmax,1000)
plot(w, f1(w), label = "$f_1(\omega)$")
plot(w, f2(w), label = "$f_2(\omega^*)$")
plot(w, 0*w,"k--")
plot(w_list,w_list*0,"bo",label="Roots of $f_1$")
plot(wstar_list,wstar_list*0,"go",label="Roots of $f_2$")
xlabel("$\omega$")
yl = 10
ylim((-yl,yl))
legend()
```

Lets compute the variances and functions (evaluated numerically within the $[a,\,a]$ interval) from these values. For convenience,
$$
\begin{array}{cc}
    \sigma_i^2 = \dfrac{2c}{\omega_i^2 + c^2} & 
    v_i(t) = \dfrac{\cos (\omega_i t)}{ \sqrt{ a + \dfrac{\sin (2\omega_i a)}{2 \omega_i }} } \\
    {\sigma_i^*}^2 = \dfrac{2c}{\omega_i^{*2} + c^2} & 
    v_i^*(t) = \dfrac{\sin (\omega_i^* t)}{ \sqrt{ a - \dfrac{\sin (2\omega_i^* a)}{2 \omega_i^* }} } 
\end{array}
$$

```python
sigma2 = 2*c/(w_list**2 + c**2)
sigma2_star = 2*c/(wstar_list**2 + c**2)
vi = lambda w,t: cos(w*t)/sqrt(a + sin(2*w*a)/(2*w))
vi_star = lambda w,t: sin(w*t)/sqrt(a - sin(2*w*a)/(2*w))


figsize(20,5)
subplot(1,2,1)
i = 1
for w in w_list:
    vi_eval = vi(w,t)
    plot(t,vi_eval, label = "$f_{"+"{}".format(i)+"}(t)$")
    i += 1
legend()
xlabel("$t$")
subplot(1,2,2)
i = 1
for w in wstar_list:
    vi_eval = vi_star(w,t)
    plot(t,vi_eval, label = "$f_{"+"{}".format(i)+"}(t)$")
    i += 1
xlabel("$t$")
legend()
```

One consequence of eigenvalue decomposition in linear algebra is that you can re-compose the original matrix using the eigenvectors and eigen values as was shown before. This is true also for eigenfunctions. We can re-build the correlation function $C(\tau)$ due to a limited number of eigenfunctions (truncation).

In the discrete case we did:

$$
C = \mathbf{v}^T \mathbf{\Sigma} \mathbf{v} 
$$

For the continuous case the correlation function can be re-constructed (or expanded, depending on how we need to view this) via 
$$
c(t_1,t_2) = \sum_{k=1}^{\infty} \Big[ \sigma_k^2 f_k(t_1) f_k(t_2) +  {\sigma_k^*}^2 f_k^*(t_1) f_k^*(t_2)  \Big]
$$


```python
T1,T2 = meshgrid(t,t)
C = 0*T1

Nterms = min(len(w_list),len(wstar_list))

for i in range(Nterms):
    if i%2 == 0: #is even
        w = w_list[i]
        sigma2 = 2*c/(w**2 + c**2)
        C += sigma2*vi(w,T1)*vi(w,T2)
    else: #odd
        w = wstar_list[i]
        sigma2 = 2*c/(w**2 + c**2)
        C += sigma2*vi_star(w,T1)*vi_star(w,T2)
```
